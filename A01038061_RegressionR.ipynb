# Regression practice in R
## Dataset: DatosGeoespaciales
Analayzing dataset and improving regression in variables

```r
#Import data set
df<-read.csv("DatosGeoespaciales.csv")
head(df)
```

## Exploratory Data Analysis

```r
#Dimensions and variables
cat("dimensions:",dim(df))
ls(df)
```

```r
summary(df)
```

```r
# Missing values
cat("missing values:", sum(is.na(df)))
```

```r
#Check the name of variables.
names(df)
```

```r
#Inspect denue_primaria with all the variables lineal regression
lmdenue_primaria = lm(denue_primaria~., data = df)
summary(lmdenue_primaria)
```

```r
#plot lm
plot(lmdenue_primaria)
```

```r
#Fitt with the strongest significative (***) and coorelation variables
#denue_preescolar, denue_secundaria,denue_departamental, ine_escuela, denue_superior
fitt_lmdenue_primaria = lm(denue_primaria~denue_preescolar+denue_secundaria
        +denue_departamental+ine_escuela+denue_superior, data = df)
summary(fitt_lmdenue_primaria)
```

### Assumptions

```r
par(mfrow=c(2,2))
plot(fitt_lmdenue_primaria)
#Homoscedasticity of residuals or equal variance
```

```r
#The regression model is linear in parameters
#The mean of residuals close to zero
mean(fitt_lmdenue_primaria$residuals)
```

```r
#No autocorrelation of residuals
library(ggplot2)
acf(fitt_lmdenue_primaria$residuals)
```

```r
#The X variables and residuals are uncorrelated
cor.test(df$denue_primaria, fitt_lmdenue_primaria$residuals)
```

```r
#variability in X values is positive
var(df$denue_primaria)
```

```r
#No perfect multicollinearity
library(car)
vif(fitt_lmdenue_primaria)
```

```r
#Check Assumptions Automatically
par(mfrow=c(2,2))
gvlma::gvlma(fitt_lmdenue_primaria)
```

## Select variables with forward

```r
#Regression without explicative variables
RLM.Empty<-lm(denue_primaria~1,data=df)
summary(RLM.Empty)
```

```r
RLM.Complete<-lm(denue_primaria~.,data=df)
summary(RLM.Complete)
```

```r
#Selection of variables with Forward selector
RLM.Forward<-step(RLM.Empty, scope=list(lower=RLM.Empty, upper=RLM.Complete), direction="forward")
summary(RLM.Forward)
```

```r
par(mfrow=c(2,2))
plot(RLM.Forward)
```

```r
plot(AIC(RLM.Forward))
```
